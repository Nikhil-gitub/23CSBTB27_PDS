```python
import pandas as pd
import matplotlib.pyplot as plt

department_budget = pd.DataFrame({
    'Department': ['HR', 'IT', 'Finance', 'Marketing', 'Operations'],
    'Budget': [50000, 120000, 80000, 60000, 100000]
})


actual_expenses = pd.DataFrame({
    'Department': ['HR', 'IT', 'Finance', 'Marketing', 'Operations'],
    'Actual_Expenses': [55000, 115000, 90000, 65000, 95000]
})


merged_df = pd.merge(department_budget, actual_expenses, on='Department')


merged_df['Variance'] = merged_df['Actual_Expenses'] - merged_df['Budget']


merged_df['Exceeding_Budget'] = merged_df['Variance'] > 0

merged_df = merged_df.sort_values(by='Variance', ascending=False)

print("Merged Data with Variance:")
print(merged_df)


plt.figure(figsize=(8,5))
plt.bar(merged_df['Department'], merged_df['Budget'], label='Budget', alpha=0.7)
plt.bar(merged_df['Department'], merged_df['Actual_Expenses'], label='Actual Expenses', alpha=0.7)
plt.xlabel("Department")
plt.ylabel("Amount")
plt.title("Department Budget vs Actual Expenses")
plt.legend()
plt.show()
```
